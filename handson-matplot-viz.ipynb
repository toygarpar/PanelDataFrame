```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#GLOBAL API
x = np.arange(-15,13) # create a range of numbers from -15 to 13
```

```python
plt.figure(figsize=(14,6)) # set the figure size

plt.title("The Plot") # set the title of plot

plt.plot(x, x**2, label="y=x^2", color='red') # plot x vs x^2
plt.plot(x, -3*x**2, label="-3x^2", color='blue') # plot x vs -3x^2
plt.plot(x, 4*x**2+6, label="4x^2+6", color='green') # plot x vs 4x^2+6
plt.legend() # show the legend on the plot
plt.grid(True) # add grid to the plot
plt.show() # display the plot
```

```python
plt.figure(figsize=(14,6))  # Create a figure with width=14 inches and height=6 inches

plt.title("The Plot 2")  # Set the title for the entire figure

# First subplot (left panel)
plt.subplot(1,2,1)  # Create a 1x2 grid of subplots and select the first one
plt.plot(x, x**2, label="y=x^2", color='red')  # Plot x vs x² with red color
plt.plot([0,0,0],[-10,0,100])  # Plot a vertical line at x=0 from y=-10 to y=100
plt.legend(["X^2", "Vertical Line"])  # Add legend with specified labels
plt.xlabel("X-axis")  # Label for x-axis
plt.ylabel("Y-axis - X Squared")  # Label for y-axis

# Second subplot (right panel)
plt.subplot(1,2,2)  # Select the second subplot in the 1x2 grid
plt.plot(x, -1 * (x ** 2))  # Plot x vs -x² (default color)
plt.plot([-10,0,10], [-50,-50,-50])  # Plot a horizontal line at y=-50 from x=-10 to x=10
plt.legend(["-X^2", "Horizontal Line"])  # Add legend with specified labels
plt.xlabel("X-axis")  # Label for x-axis
plt.ylabel("Y-axis - Neg X Squared")  # Label for y-axis

plt.show()  # Display the figure with both subplots
```

```python
# Plotting the first line: X^2
# x: The x-coordinates of the points on the plot.
# (x ** 2): The y-coordinates, which are the squares of the x-coordinates.
# color="red": Sets the color of the line to red.
# linewidth=3: Sets the width of the line to 3 units.
# marker="o": Places a circular marker at each data point on the line.
# markersize=8: Sets the size of each marker to 8 units.
# label="X^2": Provides a label for this line, which will be used in the legend.
axes.plot(
    x, (x ** 2), color="red", linewidth=3, marker="o", markersize=8, label="X^2"    
)  # Plot x vs x² with specified properties           

# Setting the x-axis label
# Sets the label of the x-axis to "X".
axes.plot( x, -1 * (x ** 2), "b--", label="-X^2") #     Plot x vs -x² with specified properties


# Setting the x-axis label
# Sets the label of the x-axis to "X".
axes.set_xlabel("X") #  Label for x-axis

# Setting the y-axis label
# Sets the label of the y-axis to "X Squared".
axes.set_ylabel("X Squared") #      Label for y-axis

# Setting the title of the plot
# Sets the title of the plot to "The Plot 3".
axes.set_title("The Plot 3") # Title of the plot


# Adding the legend to the plot
# Displays a legend that includes labels for each plotted line, helping to identify them.
axes.legend()

fig # Displaying the figure with all defined properties
```

```python
# make subplots using grid_like format with subplot2grid() function

# Create a new figure with a specific size (14 inches wide, 6 inches tall)
plt.figure(figsize=(14, 6))

# Create subplots using subplot2grid with a 3x3 grid layout
# This allows for flexible subplot arrangements with row/column spanning

# ax1: Top plot spanning all 3 columns (full width of the figure)
# Position: top row (0), starting at column 0, spanning 3 columns
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)

# ax2: Middle-left plot spanning 2 columns
# Position: middle row (1), starting at column 0, spanning 2 columns
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)

# ax3: Right-side plot spanning 2 rows (middle and bottom)
# Position: middle row (1), starting at column 2, spanning 2 rows down
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)

# ax4: Bottom-left plot (single cell)
# Position: bottom row (2), column 0
ax4 = plt.subplot2grid((3,3), (2,0))

# ax5: Bottom-center plot (single cell)
# Position: bottom row (2), column 1
ax5 = plt.subplot2grid((3,3), (2,1))

# Display the figure with all the subplots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create data for plotting
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5)
y4 = np.random.normal(0, 1, 100)
y5 = x**2

# Create figure and subplots with custom layout
plt.figure(figsize=(14, 6))

# ax1: Top plot (full width) - Sine wave
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_title('Sine Wave (Full Width)')
ax1.legend()
ax1.grid(True)

# ax2: Middle-left plot (2 columns) - Cosine wave
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax2.plot(x, y2, 'r--', label='cos(x)')
ax2.set_title('Cosine Wave')
ax2.legend()
ax2.grid(True)

# ax3: Right-side plot (2 rows tall) - Exponential decay
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax3.plot(x, y3, 'g-', label='exp(-x/5)')
ax3.set_title('Exponential Decay')
ax3.legend()
ax3.grid(True)

# ax4: Bottom-left plot - Random scatter
ax4 = plt.subplot2grid((3,3), (2,0))
ax4.scatter(x, y4, c='purple', alpha=0.6, label='Random noise')
ax4.set_title('Random Scatter')
ax4.legend()
ax4.grid(True)

# ax5: Bottom-center plot - Quadratic function
ax5 = plt.subplot2grid((3,3), (2,1))
ax5.plot(x, y5, 'm-', label='x²')
ax5.set_title('Quadratic Function')
ax5.legend()
ax5.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# HISTOGRAMS
values = np.random.randn(1000)
```

```python
# Create a new figure with specified size (12 inches wide, 6 inches tall)
plt.subplots(figsize=(12, 6))

# Create a histogram with the following parameters:
# - values: the data to be plotted (should be defined elsewhere in your code)
# - bins=100: divides the data range into 100 equal-width bins
# - alpha=0.8: sets transparency level (0=transparent, 1=opaque)
# - histtype='bar': creates a traditional bar-style histogram
# - color='steelblue': sets the fill color of the bars
# - edgecolor='green': sets the color of the bar edges
plt.hist(values, bins=100, alpha=0.8,
          histtype='bar', color='steelblue',
          edgecolor='green')

# Set the x-axis limits to range from -5 to 5
# This will clip any data outside this range
plt.xlim(xmin=-5, xmax=5)

# Display the plot
plt.show()
```

```python
fig.savefig('histogram_plot.png')
```

```python
from scipy.stats import gaussian_kde

density = gaussian_kde(values)
density # density is a function that computes the probability density 
```

```python
# Create a new figure with specified size (12 inches wide, 6 inches tall)
plt.subplots(figsize=(12, 6))

# Generate a range of x-values for plotting the density curve:
# - Starts 10 units below the minimum of 'values' 
# - Ends 10 units above the maximum of 'values'
# - Creates 100 equally spaced points in this range
values2 = np.linspace(min(values)-10, max(values)+10, 100)

# Plot the density curve:
# - values2: x-axis coordinates
# - density(values2): y-axis values (assuming 'density' is a predefined function, which we did)
# - color='#FF7F00': orange color (hex code)
plt.plot(values2, density(values2), color='#FF7F00')

# Fill the area under the density curve:
# - values2: x-axis coordinates
# - 0: baseline y-value (bottom of fill)
# - density(values2): top of fill (follows density curve)
# - alpha=0.5: semi-transparent fill (50% opacity)
# - color='#FF7F00': same orange color as the line
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')

# Set x-axis limits to focus on the range -5 to 5
# (This will clip any data/density outside this range)
plt.xlim(xmin=-5, xmax=5)

plt.title("Density Estimate")  # Add a title
plt.xlabel("Values")           # Label x-axis
plt.ylabel("Density")          # Label y-axis
plt.grid(alpha=0.3)            # Add light grid lines

# Display the plot
plt.show()
```

```python
# combine the two plots into one figure with subplots, combined histogram and density plot

# Create a new figure with specified size (12 inches wide, 6 inches tall)
plt.subplots(figsize=(12, 6))

# Plot histogram of 'values' with the following parameters:
# - values: input data (should be defined earlier)
# - bins=100: divides data into 100 bins for fine detail
# - alpha=0.8: 80% opacity (semi-transparent)
# - density=1: normalizes histogram to integrate to 1 (probability density)
# - histtype='bar': standard bar-style histogram
# - color='steelblue': fill color of bars
# - edgecolor='green': outline color of bars
plt.hist(values, bins=100, alpha=0.8, density=1,
          histtype='bar', color='steelblue',
          edgecolor='green')

# Overlay density curve:
# - values2: x-coordinates (should be defined earlier, e.g., with np.linspace)
# - density(values2): y-coordinates from density function
# - color='#FF7F00': orange color (hex code)
# - linewidth=3.0: thicker line for better visibility
plt.plot(values2, density(values2), color='#FF7F00', linewidth=3.0)

# Set x-axis limits to focus on range [-5, 5]
plt.xlim(xmin=-5, xmax=5)

# Display the plot
plt.show()
```

```python
# BAR CHART EXAMPLE

Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
# Create a new figure with specified size (12 inches wide, 4 inches tall)
plt.figure(figsize=(12, 4))

# Set the width of each bar in the plot (0.5 = half of default width)
barWidth = 0.5

# Create a bar plot with the following parameters:
# - np.arange(len(Y)): x-coordinates (automatically generates positions 0,1,2,...)
# - Y: heights of the bars (should be defined earlier as a list/array of values)
# - width=barWidth: sets bar width to 0.5 (default is 0.8)
# - color='#00b894': sets bar color to a mint green (hex code)
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894')

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 4))

barWidth = 0.5
plt.bar(np.arange(len(Y)), Y, width=barWidth, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=barWidth, color='#e17055', bottom=Y, label='Label Y2') # bottom argument is used to stack the second bar on top of the first one , label argument is used to add a legend

plt.legend()
plt.show()
```
