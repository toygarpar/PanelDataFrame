```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
# Create a connection to the SQLite database: form a connection and return a connection object

#to have a connection to the database, we need to create a connection object
conn = sqlite3.connect('chinook.db')
```

```python
# then we can create a cursor object from this connection

cursor = conn.cursor()
```

```python
# execute method on the cursor object to execute a SQL query and fetch data from the database

cursor.execute("SELECT * FROM employees LIMIT 5;")
```

```python
# to actually fetch the data, we can use the fetchall() method on the cursor object
results = cursor.fetchall()
```

```python
results
```

```python
df = pd.DataFrame(results) # convert the results into a pandas dataframe
```

```python
df
```

```python
cursor.close() # close the cursor object
conn.close() # close the connection object
# The connection to the database is now closed, and we can no longer execute SQL queries.
```

```python
# use the pandas `read_sql` function to read the results of a SQL query directly into a pandas `DataFrame`

conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql("SELECT * FROM employees LIMIT 5;", conn)
df.head()
```

```python
df = pd.read_sql("SELECT * FROM employees LIMIT 5;", conn,
                   index_col='EmployeeId', parse_dates=['BirthDate', 'HireDate']) # set the index of the DataFrame to the EmployeeId column, and parse the BirthDate and HireDate columns as dates
df.head()
```

```python
#read_sql_query
df = pd.read_sql_query("SELECT * FROM employees LIMIT 5;", conn,
                   index_col='EmployeeId', parse_dates=['BirthDate', 'HireDate']) # set the index of the DataFrame to the EmployeeId column, and parse the BirthDate and HireDate columns as dates
df.head()
```

```python
#read_sql_table, works only with SQLALchemy
from sqlalchemy import create_engine
engine = create_engine('sqlite:///chinook.db') # create an engine object to connect to the SQLite database
# the create_engine function creates a new SQLAlchemy engine instance, which is a factory for creating database connections
# the engine object is used to connect to the database and execute SQL queries
# the engine object is a high-level interface for connecting to the database and executing SQL queries


# create a connection to the database using the engine object
conn = engine.connect() # create a connection object from the engine object

# then read the data from the database using the read_sql_table function
# the read_sql_table function reads a table from the database and returns a pandas DataFrame
df = pd.read_sql_table("employees", conn, index_col='EmployeeId', parse_dates=['BirthDate', 'HireDate']) # set the index of the DataFrame to the EmployeeId column, and parse the BirthDate and HireDate columns as dates


df.head()
```

```python
df.to_sql("employees2", conn, if_exists="replace") # write the DataFrame to a new table in the database

#if_exists="replace" means that if the table already exists, it will be replaced with the new data, append means that if the table already exists, the new data will be appended to the existing data, fail means that if the table already exists, an error will be raised
```

```python
pd.read_sql_query("SELECT * FROM employees2 LIMIT 5;", conn) # read the data from the new table in the database
```
